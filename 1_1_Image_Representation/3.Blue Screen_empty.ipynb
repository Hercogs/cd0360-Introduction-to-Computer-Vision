```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

### Read in and display image

```python
# Read in the image
image = cv2.imread("images/pizza_bluescreen.jpg")

print("This image is: ", type(image),
     " whith dimensions: ", image.shape)
```

```python
%matplotlib inline

# DIsplay image
plt.imshow(image)
```

```python
# Reverse BGR to RGB
image_copy = np.copy(image)

image_copy = image[:, :, ::-1]  # or cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

### Define color threshold

```python
# Define our color selection boundaries in RGB values
lower_blue = np.array([0, 0, 230])
upper_blue = np.array([50, 50, 255])
```

```python
# Define the mask area
mask = cv2.inRange(image_copy, lower_blue, upper_blue)

# Visualize mask
plt.imshow(mask, cmap="grey")
```

```python
masked_image = np.copy(image_copy)

masked_image[mask != 0] = [0, 0, 0]
```

```python
plt.imshow(masked_image)
```

```python
# Add background to image

new_background_image = np.copy(masked_image)
new_background_image[mask != 0] = [255, 0, 0]
plt.imshow(new_background_image)
```
